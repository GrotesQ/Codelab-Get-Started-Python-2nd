```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import urllib.request

keyword = input( '검색어를 입력하세요' )
browser = webdriver.Chrome( './chromedriver' )
url = 'https://imgur.com/search?q=' + keyword
browser.get( url )

html = BeautifulSoup( browser.page_source, 'html.parser' )
anchors = html.select( '.image-list-link' )
links = []

for a in anchors :
    links.append( 'https://imgur.com' + a['href'] )

for link in links :
    try :
        browser.get( link )
        time.sleep( 1 )
        html = BeautifulSoup( browser.page_source, 'html.parser' )
        post = html.find( class_="post-image" )
        img = post.find( 'img' )
        if img is None :
            video = post.find( 'video' )
            source = video.find( 'source' )
            src = source[ 'src' ]
        else :
            src = img[ 'src' ]
        src = img[ 'src' ]
        path = 'https:' + src
        name = src.split( '/' ).pop()

        urllib.request.urlretrieve( path, './images/' + name ) # 원본 주소, 저장할 이름
        print( '다운로드 : ' + path )
    except :
        print( '다운로드 실패 : ' + link )
    
# for img in imgs :
#     src = img[ 'src' ]
#     path = 'https:' + src
#     # 'i.imgur.com/4AiXzf8b.jpg' -> split -> [ 'i.imgur.com', '4AiXzf8b.jpg' ]
#     name = src.split( '/' ).pop()
    
#     print( '다운로드 시작 : ' + path )
#     urllib.request.urlretrieve( path, './images/' + name ) # 원본 주소, 저장할 이름

browser.close()
```
