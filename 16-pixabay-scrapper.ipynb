```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

keyword = input( '검색어를 입력하세요' )
url = 'https://pixabay.com/ko/images/search/' + keyword + '/'
browser = webdriver.Chrome( './chromedriver' )
browser.get( url )

html = BeautifulSoup( browser.page_source, 'html.parser' )
imgs = html.select( '.search_results .item img' )
srcs = []

for img in imgs :
    srcs.append( img[ 'src' ] )

for src in srcs :
    try :
        browser.get( src )
        browser.save_screenshot( './images/screenshot-' + src.split('/').pop() )
    except :
        print( '다운로드 실패 : ' + src )
        
browser.close()
```
